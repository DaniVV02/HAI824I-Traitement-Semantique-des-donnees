```python
from SPARQLWrapper import SPARQLWrapper, JSON

# Set up the endpoint
endpoint_url = "https://www.imgt.org/fuseki/MabkgKg/sparql"
sparql = SPARQLWrapper(endpoint_url)

# Example query - Triple count
query = """
SELECT (COUNT(*) AS ?triples) WHERE {
    ?s ?p ?o
}
"""

sparql.setQuery(query)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

# Print results
for result in results["results"]["bindings"]:
    print(f"Total triples: {result['triples']['value']}")
```

```python
## Lister  20 mAbs ainsi que leur cibles
sparql.setQuery(
    prefix
    + """
SELECT * WHERE {
  ?target imgt:isTargetOf ?mAb
} LIMIT 20
    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(r["mAb"]["value"], "-->", r["target"]["value"])
except Exception as e:
    print(e)
```

```python
## Lister 10  mAbs et les indications cliniques associées
sparql.setQuery(
    prefix
    + """
SELECT * WHERE {
  ?mAb imgt:hasClinicalIndication ?clinicalIndication ;
        a imgt:INN.
} LIMIT 10
    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(r["mAb"]["value"], "-->", r["clinicalIndication"]["value"])
except Exception as e:
    print(e)
```

```python
## Lister 20 mAbs qui sont associées à un mode d’action
sparql.setQuery(
    prefix
    + """
SELECT distinct * WHERE {
  ?mAb bao:BAO_0000196 ?MOA
} LIMIT 20
    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(r["mAb"]["value"], "-->", r["MOA"]["value"])
except Exception as e:
    print(e)
```

```python
## Associer 30 mAbs avec leur produit d’étude
sparql.setQuery(
    prefix
    + """
SELECT distinct * WHERE {
  ?mAb imgt:hasProduct ?product .
  ?product imgt:hasStudyProduct ?study
} LIMIT 20
    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(r["mAb"]["value"], "-->", r["study"]["value"])
except Exception as e:
    print(e)
```

```python
## Rechercher des mAbs en utilisant leur nom INN, avec la propriété imgt:inn_name
sparql.setQuery(
    prefix
    + """
SELECT distinct * WHERE {
  ?mAb imgt:inn_name ?inn_name .
} LIMIT 20
    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(r["mAb"]["value"], "-->", r["inn_name"]["value"])
except Exception as e:
    print(e)
```

```python
## Suivant la même procédure rechercher les mAb avec les noms suivants: belimumab, oxelumab, relatlimab.
sparql.setQuery(
    prefix
    + """
SELECT distinct * WHERE {
  ?mAb imgt:inn_name ?inn_name .
  Filter (Contains(?inn_name,"belimumab") || Contains(?inn_name,"oxelumab")|| Contains(?inn_name,"relatlimab"))
} LIMIT 20

    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(r["mAb"]["value"], "-->", r["inn_name"]["value"])
except Exception as e:
    print(e)
```

```python
## Pour les mêmes mAbs de la question a), associé les à leur construction ainsi que les connaissances de la construction et aussi à leur cible
sparql.setQuery(
    prefix
    + """
SELECT distinct * WHERE {
  ?mAb imgt:inn_name ?inn_name ;
       imgt:hasConstruct ?construct.
    ?construct   obo:BFO_0000051 ?segment;
                imgt:hasImgtLabel ?imgtlabel;
                imgt:hasReceptorFormat ?receptformat
  Filter (Contains(?inn_name,"belimumab") || Contains(?inn_name,"oxelumab")|| Contains(?inn_name,"relatlimab"))
} LIMIT 20

    """
)

try:
    ret = sparql.queryAndConvert()
    for r in ret["results"]["bindings"]:
        print(
            r["mAb"]["value"],
            "-->",
            r["inn_name"]["value"],
            "-->",
            r["construct"]["value"],
            "-->",
            r["segment"]["value"],
            "-->",
            r["imgtlabel"]["value"],
            "-->",
            r["receptformat"]["value"],
        )
except Exception as e:
    print(e)
```
